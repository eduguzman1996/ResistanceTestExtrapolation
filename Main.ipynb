```python
#Step 1 Calculation of resistance and effective power by three dimensional extrapolation method
#Expirement developed in fresh water
g = 9.81 #[m/s^2]
Temp = 16.2 #[Celsius]

#Iterpolation of viscosity
X = [i[0] for i in table_10]
Y = [i[1] for i in table_10]
del X[0]
del Y[0]
viscosity = np.interp(Temp, X, Y)

#Iterpolatio of density
X = [i[0] for i in table_11]
Y = [i[1] for i in table_11]
del X[0]
del Y[0]
density = np.interp(Temp, X, Y)

#Table to stores the results for future estimateion of factor form
table_9 = [
    ['Vm','Fn','Fn^4 E-5','Rn E6','Ctm E-3', 'Cfom E-3','Fn^4/Cfom','Ctm/Cfom']
]
position = 0
for i in table_8[:]:
    #print(i)
    speed = i[0]
    Rtm = i[2]
    if position > 0:
        Fn = speed/sqrt(g*lwl_model)

        #Itepolation od Cfom for each value of Rn
        Rn = speed*lwl_model/(viscosity/1000000)
        #print('Rn',Rn)
        X = [i[0] for i in table_12]
        Y = [i[1] for i in table_12]
        del X[0]
        del Y[0]
        Cfom = np.interp(Rn, X, Y)
        #print('Cfom',Cfom)

        Cfom_table_12 = Cfom/1000
        Cfom_ITTC = 0.075/((log10(Rn)-2)**2)
        Cfom_ATTC = sqrt(0.242/log10(Rn*Cfom_ITTC))

        #print('Reynolds numers table ITTC ATTC',Cfom_table_12,Cfom_ITTC,Cfom_ATTC)
        Cfom = Cfom_ITTC


        #Creation of row data to be attached at iteration
        Ctm = Rtm/(0.5*density*(speed**2)*S_model)
        table = [speed,Fn,(Fn**4)*100000,Rn/1000000,Ctm*1000,Cfom*1000,(Fn**4)/(Cfom),Ctm/Cfom]
        table_9.append(table)       #Attach the new results at the end
    position = position +1
```

```python
#Plot of figure 60 to estimate the form factor
x = [i[6] for i in table_9]
y = [i[7] for i in table_9]
del x[0]
del y[0]
x = np.array(x)
y = np.array(y)

#print(x)
#print(y)

p1 = np.polyfit(x,y,1)
p2 = np.polyfit(x,y,2)
p3 = np.polyfit(x,y,3)
#print(p1)
#print(p2)
#print(p3)

plt.plot(x,y,'o')
xp = np.linspace(0,0.25,50)
plt.plot(xp,np.polyval(p1,xp),'r-')
plt.plot(xp,np.polyval(p2,xp),'b--')
plt.plot(xp,np.polyval(p3,xp),'m:')
plt.title('PLOT OF PRHASKA METHOD')
plt.xlabel('Fn^4 / Cfom')
plt.ylabel('Ctm / Cfom')

#Estimation of factor form
F_form = np.polyval(p2,0)
print('The form factor obtained with polynomial of order 2 is:',F_form)
```

```python
#Resistance estimation for 14[knots]
# values agreed by ITTC and ATTC  
viscosity = 1.1883/1000000      #[m^2/s]
density = 1.0259                #[Ton/m^3]
speed = table_8[8][1]*0.5144    #[m/s]

Cfom = table_9[8][5]/1000
Ctm = table_9[8][4]/1000
Cvm = F_form*Cfom
Cws = Ctm - Cvm

#Estimation of coefficients for the ship
Rn = speed*lwl_ship/viscosity
X = [i[0] for i in table_12]
Y = [i[1] for i in table_12]
del X[0]
del Y[0]
Cfos = np.interp(Rn, X, Y)
Cfos = Cfos/1000

Cvs = F_form*Cfos
Cts = Cws + Cvs
Ca = 0.006*(lwl_ship + 100)**(-0.16) - 0.00205

Cts = Cts + Ca

Rt = 0.5*density*(speed**2)*S_ship*Cts #Resistance in KN
    
print ('The resistacne estimation using ITTC curves for a speed of ',table_8[8][1],'[knot] is',Rt,'[kN]')
```

```python
#Table of results 
Results = [
    ['V[kts]','Cts','Resistance [KN]']
]
data = [i[1] for i in table_8]  #Speed of the ship in knots
position = 0
for i in data:
    if position > 0:
        speed = i*0.5144
        Rn = speed*lwl_ship/viscosity
        X = [i[0] for i in table_12]
        Y = [i[1] for i in table_12]
        del X[0]
        del Y[0]
        Cfos = np.interp(Rn, X, Y)
        Cfos = Cfos/1000

        #Estimation of Cvm and Cwm
        Cfom = table_9[position][5]/1000
        Ctm = table_9[position][4]/1000
        Cvm = F_form*Cfom
        Cws = Ctm - Cvm
        
        Cvs = F_form*Cfos
        Cts = Cws + Cvs
        Ca = 0.006*(lwl_ship + 100)**(-0.16) - 0.00205

        Cts = Cts + Ca

        Rt = 0.5*density*(speed**2)*S_ship*Cts #Resistance in KN
        NewRow = [i,Cts,Rt]
        Results.append(NewRow)
    position = position + 1
#print(Results)
    
```

```python
#print the curve of resitance

x = [i[0] for i in Results]
y = [i[2] for i in Results]
del x[0]
del y[0]
x = np.array(x)
y = np.array(y) 

plt.plot(x,y,'o')
xp = np.linspace(0,0.25,50)
plt.title('Ship Resistance curve  \n Three dimensional extrapolation method & using ITTC')
plt.xlabel('Velocity [m/s]')
plt.ylabel('Resistance[N]')
```
